## Data Cleaning Steps


```python
data.drop_duplicates(inplace=True)
```

```python
data.drop(columns=['WellnessCentreName'],inplace=True)
```

```python
data.columns
```

```python
data['Longitude']=data['Longitude'].astype(str).str.strip().str.replace(r'\s+','',regex=True)
```

```python
data['Longitude']=pd.to_numeric(data['Longitude'],errors='coerce')
```

```python
data['WellnessCenterNumber'].head(10)
```

```python
data['WellnessCenterNumber']=(data['WellnessCenterNumber'].astype(str).str.replace(r'[\(\)]', '',regex=True).str.replace(r'\n', ' | ',regex=True).
str.replace(r',\s*$', '',regex=True).str.strip())
```

```python
data['WellnessCenterNumber'].head(10)
```

```python
data['CityName']=data['CityName'].replace('DELHI/HQ/DIRECTORATE/MINISTRY','Delhi')
```

```python
data['WellnessCenterAddress'].fillna('Unknown',inplace=True)
```

```python
city_median_lat=data.groupby('CityName')['Latitude'].transform('median')
city_median_lon=data.groupby('CityName')['Longitude'].transform('median')
```

```python
data['Longitude']=data['Longitude'].fillna(city_median_lon)
data['Latitude']=data['Latitude'].fillna(city_median_lat)
```

```python
data['WellnessCenterCode'].fillna('Unkown',inplace=True)
```

```python
data['WellnessCenterNumber'] = data['WellnessCenterNumber'].replace('nan', np.nan)
```

```python
data['WellnessCenterNumber']=data['WellnessCenterNumber'].fillna('Not Available')
```
